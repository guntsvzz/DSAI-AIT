```python
import matplotlib.pyplot as plt
class CandyBasket():
    def __init__(self):
        self.basket = [] # automatically create an empty basket
        self.all_color = []
    
    def __len__(self):
        return len(self.basket)
    
    def __str__(self):
        return f"{[candy.color for candy in self.basket]}"
    
    # can add many candy of the same color
    def add(self, color, amount):
        if color not in self.all_color:
            self.all_color.append(color)
        for i in range(amount):
            candy_i = Candy(color)
            self.basket.append(candy_i)
        print(f"Added {amount} {color} candy")
    
    # can remove 1 candy of a specified color at a time
    def remove_candy(self, color):
        if color not in self.all_color:
            raise ValueError(f"{color} candy not in basket")
        else:
            for candy in self.basket:
                if candy.color == color:
                    self.basket.remove(candy)
                    print(f"{color} candy removed !")
                    break
                    
    # this function can transfer 1 candy at a time                
    def transfer(self, receiver, color):
        if color not in self.all_color:
            raise ValueError(f"{color} candy not in basket")
        else:
            for candy in self.basket:
                if candy.color == color:
                    self.remove_candy(color)
                    receiver.add(color, 1)
                    break
                    
    def total(self,):
        total = 0
        for candy in self.basket:
            total += candy.price
        return total
    
    # plot a bar plot to show the number of candies of each color
    def plot(self):
        counts = [0]*len(self.all_color)  # if len = 3 we get [0, 0, 0]
        for i, color in enumerate(self.all_color):
            for candy in self.basket:
                if candy.color == color:
                    counts[i] += 1
        plt.bar(self.all_color, counts, width=0.8)
        plt.show()
```

```python
# train test split the data FROM SCRATCH
def train_test_split(X, y, train_size):
    idx = np.arange(0,len(X),1)
    np.random.shuffle(idx)
    idx_train = idx[0:int(train_size*len(X))]
    idx_test = idx[len(idx_train):len(idx)]

    X_train = X[idx_train]
    X_test = X[idx_test]
    y_train = y[idx_train]
    y_test = y[idx_test]
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = train_test_split(X, y, 0.8)
```

```python
# scale your data FROM SCRATCH using Box-Cox transform, using lambda = -1
def box_cox_transform(X, lamb):
    if lamb == 0 :
        transformed_X = np.log(X)
    else :
        transformed_X = ((X**lamb)-1) / lamb
    return transformed_X

X_train = box_cox_transform(X_train, -1)
X_test = box_cox_transform(X_test, -1)
```

```python
intercept = np.ones((X_train.shape[0], 1))
X_train = np.concatenate((intercept, X_train), axis=1)
intercept = np.ones((X_test.shape[0], 1))
X_test = np.concatenate((intercept, X_test), axis=1)
```

```python
class LinearRegression:
    # if batch, set alpha to smaller values
    def __init__(self, alpha=0.0001, max_iter=10000, loss_old=np.inf, tol=1e-5, sampling_method="batch", weight_init = "normal", loss_func = 'mse', lamb =  0):
        self.alpha = alpha
        self.max_iter = max_iter
        self.loss_old = loss_old
        self.tol = tol
        self.sampling_method = sampling_method
        self.weight_init = weight_init
        self.loss_func = loss_func
        self.lamb = lamb
    
    # choice of weight initialization
    def init_weight(self, theta_size):
        if self.weight_init == 'normal':
            return np.random.randn(theta_size) # standard normal distribution
        elif self.weight_init == 'random':
            return np.random.rand(theta_size) # [0, 1)
        elif self.weight_init == 'zero':
            return np.zeros(theta_size) # [0, 1)
        else: raise ValueError("weight_init must be 'normal' , 'random' or 'zero'")
    
    def fit(self, X, y, X_test, y_test):
        self.theta = self.init_weight(X.shape[1])
        
        iter_stop = 0
        list_of_used_ix = [] # sto without replacement
        self.train_losses = []
        self.train_accuracies = []
        self.test_losses = []
        self.test_accuracies = []
        
        for i in range(self.max_iter):
            
            # choice of sampling method
            if self.sampling_method == "sto":
                i = np.random.randint(X.shape[0])
                while i in list_of_used_ix: # without replacement
                    i = np.random.randint(X.shape[0])
                X_train = X[i, :].reshape(1, -1)
                y_train = y[i]
                list_of_used_ix.append(i)
                if len(list_of_used_ix) == X.shape[0]:
                    list_of_used_ix = []
                    
            elif self.sampling_method == 'mini':
                batch_size = int(0.3 * X.shape[0])
                ix = np.random.randint(0, X.shape[0]) # mini with replacement
                X_train = X[ix:ix+batch_size]
                y_train = y[ix:ix+batch_size]
        
            elif self.sampling_method == 'batch': # batch
                X_train = X
                y_train = y
                
            else: raise ValueError("sampling_method must be 'sto' , 'mini' or 'batch'")
            
            yhat = self.h_theta(X_train)
            error = yhat - y_train
            
            # early stopping
            loss_new = self.loss_function(yhat, y_train)
            acc = self.r_square(yhat, y_train)
            self.train_losses.append(loss_new)
            self.train_accuracies.append(acc) 
            
            # Test loss and Acc
            yhat_test = self.h_theta(X_test)
            test_loss = self.loss_function(yhat_test, y_test)
            test_acc = self.r_square(yhat_test, y_test)
            self.test_losses.append(test_loss)
            self.test_accuracies.append(test_acc) 
            
            if self.delta_loss(loss_new, self.loss_old, self.tol):  #np.allclose
                iter_stop = i
                break
            self.loss_old = loss_new

            grad = self.gradient(X_train, error)
            self.theta = self.theta - self.alpha * grad

    # can name it predict for easy understanding
    def h_theta(self, X):
        return X @ self.theta
    
    # choice of loss function ** don't forget the gradient !!
    def loss_function(self, yhat, y):
        if self.loss_func == 'mse':
            return ((yhat - y)**2 / yhat.shape[0]).sum()
        elif self.loss_func == 'ridge':
            # Regularization do not penaltize the intercept so we only use theta of index 1 onwards
            return (((yhat - y)**2 / yhat.shape[0]).sum()) + (self.lamb*((self.theta[1:]**2).sum())) 
        else: raise ValueError("loss_func must be 'mse' or 'ridge'")

    def delta_loss(self, loss_new, loss_old, tol):
        return np.abs(loss_new - loss_old) < tol
    
    # choice of gradient according to the loss function
    def gradient(self, X, error):
        if self.loss_func == 'mse':
            return X.T @ error
        if self.loss_func == 'ridge':
            grad = (X.T @ error) # shape (6,1)
            ridge_grad = np.insert((self.lamb * self.theta[1:]), 0, 0) # shape (5,1) >> (6,1)
            return  grad + ridge_grad
    
    # Plot training and testing losses as number of iters increases
    def plot_loss(self):
        plt.plot(np.arange(len(self.train_losses)) , self.train_losses, label = "Train Losses")
        plt.plot(np.arange(len(self.test_losses)) , self.test_losses, label = "Test Losses")
        plt.title("Train and Test Losses")
        plt.xlabel("iteration")
        plt.ylabel("loss")
        plt.legend()
        plt.show()
        
    # Plot training and testing accuracies as number of iters increases
    def plot_acc(self):
        plt.plot(np.arange(len(self.train_accuracies)) , self.train_accuracies, label = "Train Accuracies")
        plt.plot(np.arange(len(self.test_accuracies)) , self.test_accuracies, label = "Test Accuracies")
        plt.title("Train and Test Accuracies")
        plt.xlabel("iteration")
        plt.ylabel("Accuracy")
        plt.legend()
        plt.show()
        
    #  feature importance output a bar plot, plotting the importance scores of each features in sorted order, log transform
    def feature_importance(self):
        # we do not use intercept
        name = ['LSTAT', 'INDUS', 'TAX', 'PTRATIO', 'NOX'] #change it later
        importance = np.log(np.abs(self.theta[1:])) # cannot do log on negative values so we do absolute on them first
        
        sorted_idx = np.argsort(importance)  # sort the features according to the importance
        
        # plot feature importance
        plt.bar([name[n] for n in sorted_idx], importance[sorted_idx])
        plt.show()
    
    # score' or accuracy of linear regression can be reported in the form of R^2
    # sklearn : https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression.score
    def r_square(self, yhat, y):
        return 1 - ( ((y - yhat)** 2).sum() / ( ( (y - y.mean() ) ** 2).sum() + 1e-10 ))
    
    # CV experiment comparing their cross-validation accuracy, and output their feature importance scores
    def cv(self, X_train, y_train, cv):
        foldsize = int(X_train.shape[0]/cv)
        loss_cv = np.zeros(cv)
        acc_cv = np.zeros(cv)
        
        for fold_idx, i in enumerate(list(range(0, X_train.shape[0], foldsize))[:-1]):
            X_val_ = X_train[i:i+foldsize]
            y_val_ = y_train[i:i+foldsize]
            
            X_train_ = np.concatenate((X_train[:i], X_train[i+foldsize:]))
            y_train_ = np.concatenate((y_train[:i], y_train[i+foldsize:]))
            
            self.fit(X_train_, y_train_, X_val_, y_val_)
            
            yhat_ = self.h_theta(X_val_)
            loss_ = self.loss_function(yhat_, y_val_)
            acc_ = self.r_square(yhat_, y_val_)
            loss_cv[fold_idx] = loss_
            acc_cv[fold_idx] = acc_
        return loss_cv.mean(), acc_cv.mean() # return average loss and accuracy from all fold
```
