#### 3.1 Dealing with missing values
- Case 1: **Continuous values**
    - Repleace with:
        - 1. **mean/median** if the distribution is normal
        - 2. **median** if the distribution is NOT normal
        - 3. **0** if 0 is every commmon occurence in the real-world
        - 4. **regression** if your data fits nicely with regression, especaially linear regression
    - Do we ever drop?
        - almost never, drop columns lose a lot of values, dropping rows also lost other features...
        - If you are super sure this is human mistake, you can drop
- Case 2: **Discrete values**
    - Replace with:
        - 1. **Majority** - when majority is like 98%
        - 2. **No category** - when we are not so sure
        - 3. **Ratio** - preserve the shape of distrubution

#### Theory/principle how tocenvert string to integer/float
Three ways:
1. **Label encoding** : Just change to number:
    1. Example: change developing to 0, developed to 1
    2. Pros: super simple
    3. Cons: imagine you got encoded 0, 1, 2, 3, 4, 5
        1. You create a unwanted of order within this, e.g. 5>1
        2. NEVER USE label encodiung uf you have more than two categories
2. **One-hot encoding** Pivot the values and make it become the columns
    1. Delete the column "Status"
    2. Put Developing as one column Developed as another column
        1. if Developing is true, put 1 and then Developed column will be 0
        2. if Developing is false, put 0 and then in the Developed column will be 1
3. **Sparse one-hot encoding** : if you have n columns of pivot, you can always delete 1 to save space.


```python
##we want you convert status to 0 and 1
#easiest way is to use sk.learn.preprocessing import LabelEncoder

#1. import the library
from sklearn.preprocessing import LabelEncoder

#2. create the object
le = LabelEncoder()

#3. fit and transform
some_make_up_data = np.array(["Male", "Female", "Female","Male","Male", "Female"]) #===one column of df
result = le.fit_transform(some_make_up_data)

print(result)

#if you want to know which map to which, use le.inverse_transform(np.array([0,1]))
'''
le.fit_transform(some_make_up_data) 
=== 
le.fit(some_make_up_data)  #learn the mapping 
result = le.transform(some_make_up_data)
'''

#Exercise: please transform status using LabelEncoder
df.status = le.fit_transform(df.status)

#to know that what map to what
le.inverse_transform(np.array([0,1]))

#test that my transform is correct
assert len(df.status.unique()) == 2
#you can try something df.status.unique()) == [1.0]

#i want to tell the difference between fit snd transform
#fit mean learn the pattern
#transform mean transform the data
```

#### 3.3 Standardization / Normalization

- i.e., we usually do splitting before any preprocessing
- splitting - test train split before any preprocessing
    - if you standardize 
    - if you fill missing value with some global statistics, you should split
    - (it's ok to label encode)
- when some knowleadge of the testing set flows to the training set, we call "data leakage" --> this is one of MOST COMMON MISTAKE of all beginners... like 99%

```python
data = np.array([[1,5],[2,6],[3,7],[4,8]])

np.array([1,2,3,4]).std()
```

```python
#for each feature, do (x-mean)/std
#for the first feeature, mean = (1+2+3+4)/4 = 2.5
#   for the first data, do (1-2.5)/1.11 = -1.35
#   for the second data, do (2-2.5)/1.11 = -0.45

#after standardization, all features will have same scale
#all features mean = 0, std = 1
 
#1. import the library
from sklearn.preprocessing import StandardScaler

#2. create the object
sc = StandardScaler()
'''
The standard score of a sample x is calculated as: z = (x - u) / s
'''
#3. fit and transform
standardized_data = sc.fit_transform(data)
print(sc.mean_) #(1+2+3+4)/4 = 2.5 #(5+6+7+8)/4 = 6.5
print(standardized_data)

#when do you want to standardize your data:
#1. when you use regression, definitely yes - feautres importance
#2. in most classification algorithms (not distance-based algorithm), 
#   you benefit a bit from standardization
#3. your training will be more stable

#when you should NOT standardize data
#1. when the mean does not represent well your data
#   signal
#we can either do nothing or domin-max normalization
#from sklearn.preprocessing import MinMaxScaler #try this at your home
# (x-xmax)/(xmax-xmin) ==> (0,1)
#2. 
```
