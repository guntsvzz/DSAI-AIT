```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu_1 = np.array([10,10])
mu_2 = np.array([20,10])
mu_3 = np.array([20,20])
sigma = np.array([[4,0],[0,4]])
X1 = np.random.multivariate_normal(mu_1, sigma,100)
X2 = np.random.multivariate_normal(mu_2, sigma,100)
X3 = np.random.multivariate_normal(mu_3, sigma,100)
X = np.concatenate((X1,X2,X3),axis=0)

plt.scatter(X[:,0],X[:,1])
plt.title("Input Data")
plt.show()
```

```python
n_sample = 1000
data_0 = np.random.multivariate_normal(mean=[-1, 1], cov=[[2, 1],[1, 2]], size=(n_sample))
data_1 = np.random.multivariate_normal(mean=[1, 5], cov=[[3, 1],[2, 1]], size=(n_sample))
data_2 = np.random.uniform([5, -3], [10, 5], size=(n_sample, 2))
X = np.concatenate((data_0,data_1,data_2),axis=0)
X.shape
```

```python
plt.scatter(X[:,0],X[:,1])
plt.title("Input Data")
plt.show()
```

```python
iters = 10
m,n = X.shape
k = 3 
indices = np.arange(m)
np.random.shuffle(indices)

#1. Initialize means
means = X[indices[0:3],:]
means

#2. Repeat until convergence

C = np.zeros(m,dtype=int)
for iter in range(iters):
    plt.scatter(X[:,0],X[:,1])
    plt.scatter(means[:,0],means[:,1],c ='red',s=100)
    plt.title('Input Data')
    plt.show()
    #2.1 map each datum to its nearest cluster center

    dist = 0
    for i in range(m):
        C[i] = np.argmin(np.linalg.norm(X[i,:]-means,axis=1))
        dist = dist + np.linalg.norm(X[i,:]-means[C[i],:])**2
    dist = dist/m
    print('Iter %d : distortion = %f' % (iter,dist))
    
    #2.2 adjust means
    for j in range(k):
        means[j,:]=np.mean(X[C==j],axis=0)

```
