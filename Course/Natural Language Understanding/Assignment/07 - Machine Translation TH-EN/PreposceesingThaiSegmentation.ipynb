## 1. Thai Word Segmenter

```python
from pythainlp.tokenize import word_tokenize
text = "โอเคบ่ถ้าพวกเรารักภาษาบ้านเกิด"
text = 'จังหวัดเชียงรายตั้งอยู่ตอนเหนือสุดของประเทศไทย'
```

```python
word_tokenize(text, engine="newmm")
```

```python
word_tokenize(text, engine="deepcut") 
```

```python
word_tokenize(text, engine="longest")
```

```python
word_tokenize(text, engine="mm")
```

```python
from attacut import tokenize, Tokenizer
atta = Tokenizer(model="attacut-c")
words = atta.tokenize(text)
words
```

```python
atta = Tokenizer(model="attacut-sc")
words = atta.tokenize(text)
words
```
