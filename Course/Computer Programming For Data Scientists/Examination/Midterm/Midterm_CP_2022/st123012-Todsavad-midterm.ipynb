````python
class LogisticRegression:
    # if batch, set alpha to smaller values
    def __init__(self, alpha=0.0001, max_iter=10000, loss_old=np.inf, tol=1e-5, sampling_method="batch", weight_init = "normal", loss_func = 'bce', lamb =  0):
        self.alpha = alpha
        self.max_iter = max_iter
        self.loss_old = loss_old
        self.tol = tol
        self.sampling_method = sampling_method
        self.weight_init = weight_init
        self.loss_func = loss_func
        self.lamb = lamb
    
    # choice of weight initialization
    def init_weight(self, theta_size):
        if self.weight_init == 'normal':
            return np.random.randn(theta_size) # standard normal distribution
        elif self.weight_init == 'random':
            return np.random.rand(theta_size) # [0, 1)
        elif self.weight_init == 'zero':
            return np.zeros(theta_size)
        elif self.weight_init == 'one':
            return np.ones(theta_size) 
    
    def fit(self, X, y, X_test, y_test):
        self.theta = self.init_weight(X.shape[1])
        
        iter_stop = 0
        list_of_used_ix = [] # sto without replacement
        self.train_losses = []
        self.train_accuracies = []
        self.test_losses = []
        self.test_accuracies = []
        
        for i in range(self.max_iter):
            
            # choice of sampling method
            if self.sampling_method == "sto":
                i = np.random.randint(X.shape[0])
                while i in list_of_used_ix: # without replacement
                    i = np.random.randint(X.shape[0])
                X_train = X[i, :].reshape(1, -1)
                y_train = y[i]
                list_of_used_ix.append(i)
                if len(list_of_used_ix) == X.shape[0]:
                    list_of_used_ix = []
                    
            elif self.sampling_method == 'mini':
                batch_size = int(0.3 * X.shape[0])
                ix = np.random.randint(0, X.shape[0]) # mini with replacement
                X_train = X[ix:ix+batch_size]
                y_train = y[ix:ix+batch_size]
        
            elif self.sampling_method == 'batch': # batch
                X_train = X
                y_train = y
                
            else: raise ValueError("sampling_method must be 'sto' , 'mini' or 'batch'")
            
            yhat = self._h_theta(X_train)
            error = yhat - y_train
            
            # early stopping
            loss_new = self.loss_function(yhat, y_train)
            acc = self.r_square(yhat, y_train)
            self.train_losses.append(loss_new)
            self.train_accuracies.append(acc) 
            
            # Test loss and Acc
            yhat_test = self._h_theta(X_test)
            test_loss = self.loss_function(yhat_test, y_test)
            test_acc = self.r_square(yhat_test, y_test)
            self.test_losses.append(test_loss)
            self.test_accuracies.append(test_acc) 
            
            if self.delta_loss(loss_new, self.loss_old, self.tol):  #np.allclose
                iter_stop = i
                break
            self.loss_old = loss_new

            grad = self.gradient(X_train, error)
            self.theta = self.theta - self.alpha * grad

    # can name it predict for easy understanding
    def h_theta(self,X): #fx, h_theta 
        return self.sigmoid(X @ self.theta)

    def _h_theta(self,X): #return either 0 or 1
        return np.round(self.sigmoid(X @ self.theta))

    def sigmoid(self,h):
        return 1/(1 + np.exp(-h))

    def binary_cross_entopy(self,yhat,y):
        return -np.sum(y*np.log(yhat)+(1-y)*(np.log(1-yhat)))

    # choice of loss function 
    def loss_function(self, yhat, y):
        if self.loss_func == 'bce': #binary_cross_entopy
            return (-np.sum(y*np.log(yhat)+(1-y)*(np.log(1-yhat))))
        elif self.loss_func == 'ridge': 
            return (-np.sum(y*np.log(yhat)+(1-y)*(np.log(1-yhat)))) + (self.lamb*((self.theta[1:]**2).sum())) 
        elif self.loss_func == 'lasso':
            return (-np.sum(y*np.log(yhat)+(1-y)*(np.log(1-yhat)))) + (self.lamb*((np.abs(self.theta[1:])).sum())) 

    def delta_loss(self, loss_new, loss_old, tol):
        return np.abs(loss_new - loss_old) < tol

    def gradient(self, X, error):
        if self.loss_func == 'bce':
            return X.T @ error
        if self.loss_func == 'ridge':
            grad = (X.T @ error) # shape (6,1)
            ridge_grad = np.insert((self.lamb * self.theta[1:]), 0, 0)
            return  grad + ridge_grad
        if self.loss_func == 'lasso':
            grad = (X.T @ error) # shape (6,1)
            ridge_grad = np.insert((self.lamb * self.theta[1:]), 0, 0) 
            return  grad + ridge_grad
    
    def accuracy_score(self, yhat, y):
        correct = 0
        for i in range(len(y)):
            if y[i] == yhat[i]:
                correct += 1
        return correct / float(len(y)) * 100.0
        
    #  feature importance output a bar plot, plotting the importance scores of each features in sorted order, log transform
    def feature_importance(self):
        # we do not use intercept
        name = ['Total day minutes','Total day calls', 'Total day charge','Customer service calls']
        importance = np.log(np.abs(self.theta[:])) # cannot do log on negative values so we do absolute on them first
        
        sorted_idx = np.argsort(importance)  # sort the features according to the importance
        
        # plot feature importance
        plt.bar([name[n] for n in sorted_idx], importance[sorted_idx])
        plt.show()
    
    def r_square(self, yhat, y):
        return 1 - ( ((y - yhat)** 2).sum() / ( ( (y - y.mean() ) ** 2).sum() + 1e-10))
    
    # CV experiment comparing their cross-validation accuracy, and output their feature importance scores
    def cv(self, X_train, y_train, cv):
        foldsize = int(X_train.shape[0]/cv)
        loss_cv = np.zeros(cv)
        acc_cv = np.zeros(cv)
        
        for fold_idx, i in enumerate(list(range(0, X_train.shape[0], foldsize))[:-1]):
            X_val_ = X_train[i:i+foldsize]
            y_val_ = y_train[i:i+foldsize]
            
            X_train_ = np.concatenate((X_train[:i], X_train[i+foldsize:]))
            y_train_ = np.concatenate((y_train[:i], y_train[i+foldsize:]))
            
            self.fit(X_train_, y_train_, X_val_, y_val_)
            
            yhat_ = self.h_theta(X_val_)
            loss_ = self.loss_function(yhat_, y_val_)
            acc_ = self.r_square(yhat_, y_val_)
            loss_cv[fold_idx] = loss_
            acc_cv[fold_idx] = acc_
            
        return loss_cv.mean(), acc_cv.mean() # return average loss and accuracy from all fold
````

Write Pearson's correlation function from scratch and compute the correlation matrix of the given matrix. Compare your result with `numpy.corrcoef()`

$$ \rho_{X,Y} = \frac{\text{cov}(X,Y)}{\sigma_X \sigma_Y} $$

$\text{cov}$ is the covariance function

$\sigma_X$ is the standard deviation of $X$

$\sigma_Y$ is the standard deviation of $Y$

$\rho_{X,Y}$ is the pearson's correlation of $X$ and $Y$

Correlation Matrix ($R$)

$ R_{i,j} = \rho_{X_{i}, X_{j}} $

$X_{i}$ is the column $i^{th}$ of matrix $X$

$X_{j}$ is the column $j^{th}$ of matrix $X$

$R_{i,j}$ is the correlation between $X_i$ and $X_j$

````python
import numpy as np
data = np.array([[1,0,10,5],
    [2,1,5,4],
    [3,0,100,3],
    [4,1,4,2],
    [5,0,0,1]])
    
np.corrcoef(data, rowvar=False)
````

````python
def cov_value(x,y):
    mean_x = sum(x) / float(len(x))
    mean_y = sum(y) / float(len(y))
    sub_x = [i - mean_x for i in x]
    sub_y = [i - mean_y for i in y]
    sum_value = sum([sub_y[i]*sub_x[i] for i in range(len(x))])
    denominate = float(len(x)-1)
    cov = sum_value/denominate
    return cov
    
def check_vectors(array):
    length = len(array[0])
    x=[1 for a in array if len(a)!=length]

def covariance(array):
    c = [[cov_value(a,b) for a in array] for b in array]
    return c

def STD(x):
    import math
    mean_x = sum(x) / len(x)   # mean
    var  = sum(pow(x-mean_x,2) for x in x) / len(x)  # variance
    std  = math.sqrt(var) 
    return std
````
